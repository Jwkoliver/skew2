```python
from matplotlib import pyplot as plt
from Bio import SeqIO
import pandas as pd

def MinimumSkew(g):
    skewC,skewT,ansC,ansT = 0,0,[0],[0]
    N = len(g)
    for i in range(len(g)):
        if g[i] == 'C':
            skewC -= 1 
        if g[i] == 'G':
            skewC += 1
        if g[i] == 'T':
            skewT -= 1 
        if g[i] == 'A':
            skewT += 1
        ansC.append(skewC)
        ansT.append(skewT)
        if i%(N/10) == 0:
            print('{}%'.format(int((float(i)/float(N))*100))) 
    d = {'skewC':ansC, 'skewT':ansT}
    dat = pd.DataFrame.from_dict(d)
    if i==N-1:
        print("Complete   #data is stored in a Pandas Dataframe called 'data'")
        print("column1_title: 'skewC' column2_title: 'skewT'")
    return(data)

for record in SeqIO.parse("MED4.fasta", 'fasta'):
    data = MinimumSkew(str(record.seq))
```

```python
# Plotting the data with MatPlotLib

plt.plot(data['skewC'], 'r')
plt.title('Skew -C +G')
plt.show()
plt.plot(data['skewT'], 'b')
plt.title('Skew -T +A')
plt.show()
plt.plot(data['skewC'],'r')
plt.plot(data['skewT'],'b')
plt.title('Skew -C +G (Red) and -T +A (Blue) together')
plt.show()
```
